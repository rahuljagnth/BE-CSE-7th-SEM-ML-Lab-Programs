```python
import numpy as np
import matplotlib.pyplot as plt
X = np.linspace(-3,3,1000)                      #X interval=(-3,3), entries in X=1000
y = np.sin(X)
X+=np.random.normal(scale=0.05,size=1000)     #Adding randomness to X
```

```python
plt.scatter(X,y,alpha=0.3)                    #alpha=transparency
```

```python
def local_regression(x0,X,Y,tau):                              #Adding bias term to input
    x0 = np.r_[1,x0]                                             #These 2 lines are for Appending bias term
    X = np.c_[np.ones(len(X)),X]                                 #X & Y -previous inputs; tau-parameter
    
    xw = X.T *radial_kernal(x0,X,tau)
    beta = np.linalg.pinv(xw@X)@xw@Y
    
    return x0@beta

def radial_kernal(x0,X,tau):
    return np.exp(np.sum((X-x0)**2,axis=1)/(-2*tau**2))
```

```python
def plot_lwr(tau):                                                      #Graph Plot
    domain = np.linspace(-3,3,num=300)                                    #Create new with 300
    prediction = [local_regression(x0,X,y,tau) for x0 in domain]
    plt.scatter(X,y,alpha=0.3)
    plt.plot(domain,prediction,color="red")
    return plt

plot_lwr(0.04)                                                         #User input- tau value
```
